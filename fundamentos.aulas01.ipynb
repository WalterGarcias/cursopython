### Aula 01: Como juntas vários tipos de váriaveis em um mesmo texto

```python
faturamento = 2000
custo =  500
lucro = faturamento - custo

print('O faturamento da loja foi ' + str(faturamento) + ' reais.' + ' O custo da loja foi ' + str(custo) + ' reais.' + ' O lucro final da loja foi de ' + str(lucro) + ' reais.')
```

### Aula 02: Como juntar vários tipos de variáveis em um mesmo texto de uma forma mais amigável. 

```python
faturamento = 2000
custo = 500
lucro = faturamento - custo

print('O faturamento da loja no mês de Maio foi de {} reais. O custo da loja no mesmo mês foi de {} reais e por fim, o lucro total da loja foi de {} reais'
      .format(faturamento, custo, lucro))
```

### Aula 03: Fstring e mudanças de tipo de variáveis. 

```python
faturamento = 1000
custo = 500
lucro = faturamento - custo

print(f'O valor do faturamento é de {faturamento} reais e o valor total do lucro é de {lucro}')
```

### Aula 03: Formantado tipo de variável informada pela usuário

```python
faturamento = float(input('digite o valor total do faturamento'))
custo = float(input('informe o valor do custo semanal'))

print(faturamento - custo)
```
